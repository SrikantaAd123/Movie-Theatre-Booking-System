```python

seats = 350
total_bookings = 0
tickets_sold = 0
rejected_bookings = 0

while seats > 0:
    tickets = int(input("Enter number of tickets (0 to exit): "))
    if tickets == 0:
        break
    if tickets < 1 or tickets > 15:
        print("Invalid number of tickets! (1–15 only)\n")
        continue

    ages = []
    for i in range(tickets):
        age = int(input(f"Enter age of person {i+1}: "))
        ages.append(age)

    if any(age < 12 for age in ages):
        print("BOOKING REJECTED - Age restriction\n")
        rejected_bookings += 1
        continue

    if seats >= tickets:
        seats -= tickets
        total_bookings += 1
        tickets_sold += tickets
        print(f"BOOKING CONFIRMED - {tickets} tickets\n")
    else:
        print("BOOKING REJECTED - Not enough seats\n")
        rejected_bookings += 1
        break

print(f"Total Bookings: {total_bookings} | Total tickets sold: {tickets_sold} | Rejected Bookings: {rejected_bookings} | Remaining Seats: {seats}")
```

```python
def ticket_system():
    seats = 350
    total_bookings = 0
    tickets_sold = 0
    rejected_bookings = 0

    while seats > 0:
        try:
            user_input = input("Enter number of tickets (0 to exit): ")
            if not user_input: continue
            tickets = int(user_input)
        except ValueError:
            print("Invalid input! Please enter a number.\n")
            continue

        if tickets == 0:
            break

        if tickets < 1 or tickets > 15:
            print("Invalid number of tickets! (1–15 only)\n")
            continue

        # Age Check Requirement
        ages = []
        for i in range(tickets):
            age = int(input(f"Enter age of person {i+1}: "))
            ages.append(age)

        if any(age < 12 for age in ages):
            print("BOOKING REJECTED - Age restriction\n")
            rejected_bookings += 1
            continue

        if seats >= tickets:
            seats -= tickets
            total_bookings += 1
            tickets_sold += tickets
            print(f"BOOKING CONFIRMED - {tickets} tickets\n")
        else:
            print("BOOKING REJECTED - Not enough seats\n")
            rejected_bookings += 1

    print(f"Total Bookings: {total_bookings} | Total tickets sold: {tickets_sold} | Rejected Bookings: {rejected_bookings} | Remaining Seats: {seats}")

# Run the system
# ticket_system()
```
